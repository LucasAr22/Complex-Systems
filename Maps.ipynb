## Sistemas Complexos - Mapas e pontos de equilíbrio 

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Funções e intervalos

```python
def map1(mi,xt):
    return mi + xt*(1-xt)
def map2(mi,xt):
    return xt*(1+mi-xt)
def map3(mi,xt):
    return xt*(1+mi-xt**2)
def maplog(r,xt):
    return r*xt*(1-xt)
def mapbar(mi,xt):
    if xt >= 0 and xt < 0.5:
        return mi*xt
    else:
        return mi*(1-xt)
```

```python
def teia(x0, mi,func, nmax):
    nmax = 200
    px, py = np.empty((2,nmax+1,2))
    px[0], py[0] = x0, 0
    for n in range(1, nmax, 2):
        px[n] = px[n-1]
        py[n] = func(mi,px[n-1])
        px[n+1] = py[n]
        py[n+1] = py[n]
    return px,py
```

```python
x = np.linspace(-1,1,21)
x2 = np.linspace(0,1,21)
```

#### Estudo de caso para diferentes valores de $\mu$ 

```python
## style matplotlib
plt.style.use('bmh')

## MAP 1
for mi in np.linspace(0.1,0.5,5):
    plt.plot(x,map1(mi,x), linewidth = 2, label = '$\mu =$'+str(np.around(mi,decimals=2)))
plt.plot(x,x, color = 'black', linewidth = 2.5)
plt.legend()
plt.show()
```

```python
for mi in np.linspace(0.2,1,5):
    plt.plot(x,map2(mi,x), linewidth = 2, label = '$\mu=$'+str(np.around(mi,decimals=2)))
plt.plot(x,x, color = 'black', linewidth = 2.5)
plt.legend()
plt.show()
```

```python
for mi in np.linspace(0.1,1,5):
    plt.plot(x,map3(mi,x), linewidth = 2, label = '$\mu=$'+str(np.around(mi,decimals=2)))
plt.plot(x,x, color = 'black', linewidth = 2.5)
plt.legend()
plt.show()
```

#### Estabilidade 

```python
x = np.linspace(-0.4,0.4,21)
mi = 0.16
plt.title("$\mu = 0.16$")
plt.plot(x,map1(mi,x),linewidth = 2.5,label = 'map 1')
plt.scatter([-0.4,0.4],[-0.4,0.4],color = 'black', label = 'pt. eq.')
plt.plot(x,x,color = 'b',label = 'f. afim')
plt.legend()
plt.show()
```

```python
x = np.linspace(-0.8,0.8,21)
mi = 0.64
plt.title("$\mu =$"+str(mi))
plt.plot(x,map2(mi,x),linewidth = 2.5,label = 'map 2')
plt.scatter([0,0.64],[0,0.64],color = 'black', label = 'pt. eq.')
plt.plot(x,x,color = 'b',label = 'f. afim')
plt.legend()
plt.show()
```

```python
x = np.linspace(-0.8,0.8,21)
mi = 0.49
plt.title("$\mu = $"+str(mi))
plt.plot(x,map3(mi,x),linewidth = 2.5,label = 'map 3')
plt.scatter([-0.7,0,0.7],[-0.7,0,0.7],color = 'black', label = 'pt. eq.')
plt.plot(x,x,color = 'b',label = 'f. afim')
plt.legend()
plt.show()
```

### Gráfico Teia 

```python
x = np.linspace(-0.4,0.4,21)
mi = 0.16
x0 = -0.2
px,py = teia(x0,mi,map1,300)
plt.title("$\mu = 0.16$")
plt.plot(x,map1(mi,x),linewidth = 2.5,label = 'map 1')
plt.scatter([-0.4,0.4],[-0.4,0.4],color = 'black', label = 'pt. eq.')
plt.plot(x,x,color = 'b',label = 'f. afim')
plt.plot(px,py, color = 'black')
plt.legend()
plt.show()
```

```python
x = np.linspace(-0.8,0.8,21)
mi = 0.64
x0 = 0.2
px,py = teia(x0,mi,map2,300)
plt.title("$\mu =$"+str(mi))
plt.plot(x,map2(mi,x),linewidth = 2.5,label = 'map 2')
plt.scatter([0,0.64],[0,0.64],color = 'black', label = 'pt. eq.')
plt.plot(x,x,color = 'b',label = 'f. afim')
plt.plot(px,py, color = 'black')
plt.legend()
plt.show()
```

```python
x = np.linspace(-0.8,0.8,21)
mi = 0.49
x0 = 0.2
px,py = teia(x0,mi,map3,300)
plt.title("$\mu = $"+str(mi))
plt.plot(x,map3(mi,x),linewidth = 2.5,label = 'map 3')
plt.scatter([-0.7,0,0.7],[-0.7,0,0.7],color = 'black', label = 'pt. eq.')
plt.plot(x,x,color = 'b',label = 'f. afim')
plt.plot(px,py, color = 'black', label = 'Teia')
plt.legend()
plt.show()
```
